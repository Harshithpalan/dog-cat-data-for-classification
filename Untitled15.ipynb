```python
#import the library
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import cv2
import seaborn as sns
from sklearn.model_selection import train_test_split
import zipfile
from sklearn.linear_model import LogisticRegression
```

```python
#path of zipfile
zip_path = '/content/dog-cat-data-for-classification.zip'
```

```python
extract_path = '/content/dog-cat-data'
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
print("Extracted to",extract_path)
```

```python
dog_sample_path=os.path.join(extract_path,'dataset','dog-samples')
cat_sample_path=os.path.join(extract_path,'dataset','cat-samples')
```

```python
dog_images=[f for f in os.listdir(dog_sample_path)if f.endswith(('.jpg','.jpeg','.png'))]
cat_images=[f for f in os.listdir(cat_sample_path)if f.endswith(('.jpg','.jpeg','.png'))]
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random
print("sample dog image")
plt.figure(figsize=(10,5))
for i in range(min(5,len(dog_images))):
    img_path=os.path.join(dog_sample_path,random.choice(dog_images))
    img=mpimg.imread(img_path)
    plt.subplot(1,5,i+1)
    plt.imshow(img)
    plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random
print("sample cat image")
plt.figure(figsize=(10,5))
for i in range(min(5,len(cat_images))):
    img_path=os.path.join(cat_sample_path,random.choice(cat_images))
    img=mpimg.imread(img_path)
    plt.subplot(1,5,i+1)
    plt.imshow(img)
    plt.axis('off')
plt.show()
```

```python
img_size=64
data=[]
labels=[]
```

```python
for img_file in cat_images:
    img_path=os.path.join(cat_sample_path,img_file)
    img=cv2.imread(img_path)
    if img is not None:
        img=cv2.resize(img,(img_size,img_size))
        data.append(img.flatten())
        labels.append(0)
```

```python
for img_file in dog_images:
    img_path=os.path.join(dog_sample_path,img_file)
    img=cv2.imread(img_path)
    if img is not None:
        img=cv2.resize(img,(img_size,img_size))
        data.append(img.flatten())
        labels.append(0)
```

```python
for img_file in dog_images:
    img_path=os.path.join(dog_sample_path,img_file)
    img=cv2.imread(img_path)
    if img is not None:
        img=cv2.resize(img,(img_size,img_size))
        data.append(img.flatten())
        labels.append(1)
```

```python
x=np.array(data,dtype=np.float32)/255.0
y=np.array(labels)
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred=model.predict(X_test)
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test,y_pred))
```
